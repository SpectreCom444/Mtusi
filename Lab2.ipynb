```python
import pandas
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
dataframe = pandas.read_csv("C:/Users/Viktor/Documents/sonar.csv", header=None)
dataset = dataframe.values
```

```python
X = dataset[:,0:60].astype(float)
Y = dataset[:,60]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)

model = Sequential()
model.add(Dense(60, input_dim=60, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
history=model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)

fig, ax = plt.subplots(nrows=2, ncols=1)
ax[0].set_ylabel("Accuracy");
ax[1].set_xlabel("Generation");
ax[1].set_ylabel("Accuracy");
ax[0].scatter([i for i in range(len(history.history["accuracy"]))], history.history["accuracy"], alpha=0.3, marker = '.',s=70, color = 'lightskyblue')           
ax[1].scatter([i for i in range(len(history.history["loss"]))], history.history["loss"], alpha=0.3, marker = '.',s=70, color = 'red') 
```

```python
#уменьшим набор входных данных в 2 раза
X = dataset[:,0:30].astype(float)
Y = dataset[:,60]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)

model = Sequential()
model.add(Dense(30, input_dim=30, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
history=model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)

fig, ax = plt.subplots(nrows=2, ncols=1)
ax[0].set_ylabel("Accuracy");
ax[1].set_xlabel("Generation");
ax[1].set_ylabel("Accuracy");
ax[0].scatter([i for i in range(len(history.history["accuracy"]))], history.history["accuracy"], alpha=0.3, marker = '.',s=70, color = 'lightskyblue')           
ax[1].scatter([i for i in range(len(history.history["loss"]))], history.history["loss"], alpha=0.3, marker = '.',s=70, color = 'red') 
```

Как мы видим по графикам после уменьшение количества входных данных точность при обучение снизилась

```python
# добавим скрытый слой из 15 неиронов
X = dataset[:,0:30].astype(float)
Y = dataset[:,60]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)

model = Sequential()
model.add(Dense(30, input_dim=30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
history=model.fit(X, encoded_Y, epochs=100, batch_size=10, validation_split=0.1)

fig, ax = plt.subplots(nrows=2, ncols=1)
ax[0].set_ylabel("Accuracy");
ax[1].set_xlabel("Generation");
ax[1].set_ylabel("Accuracy");
ax[0].scatter([i for i in range(len(history.history["accuracy"]))], history.history["accuracy"], alpha=0.3, marker = '.',s=70, color = 'lightskyblue')           
ax[1].scatter([i for i in range(len(history.history["loss"]))], history.history["loss"], alpha=0.3, marker = '.',s=70, color = 'red') 
```

При этом, при использование дополнительного "скрытого" слоя мы можем видеть, что точность при обучение значительно повысилась, что позволяет компенсировать использование только 30 входных параметров и даже превзойти точность обучения на 60 входных параметрах без скрытого слоя

```python
X = dataset[:,0:30].astype(float)
Y = dataset[:,60]
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
```

```python
count=10

epochs=100
batch_size=10

number_of_layers=[1,2,3]
number_of_neurons=[5,10,15,20,25]


for m in range(len(number_of_layers)):
    for n in range(len(number_of_neurons)):
        total_neurons=number_of_neurons[n]*number_of_layers[m]
        fig, ax = plt.subplots(nrows=2, ncols=1)
        ax[0].set_ylabel("Accuracy");
        ax[1].set_xlabel("Generation");
        ax[1].set_ylabel("Accuracy");
        accuracy_mean=0;
        for i in range(count):

            model = Sequential()
            model.add(Dense(30, input_dim=30, activation='relu'))
            for j in range(number_of_layers[m]):
                model.add(Dense(number_of_neurons[n], activation='relu'))
            model.add(Dense(1, activation='sigmoid'))
            model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
            history =model.fit(X, encoded_Y, epochs=epochs, batch_size=batch_size,verbose = 0)
            
            accuracy_mean+=sum([i for i in history.history["accuracy"][-10:-1]])/9
                
            ax[0].scatter([i for i in range(len(history.history["accuracy"]))], history.history["accuracy"], alpha=0.3, marker = '.',s=70, color = 'lightskyblue')           
            ax[1].scatter([i for i in range(len(history.history["loss"]))], history.history["loss"], alpha=0.3, marker = '.',s=70, color = 'red') 
            
            print(f"Done number of layers: {m}/{len(number_of_layers)} ["+"="*m+" "*(len(number_of_layers)-m)+"]")
            print(f"Done number of neurons: {n}/{len(number_of_neurons)} ["+"="*n+" "*(len(number_of_neurons)-n)+"]")
            print(f"Done: {i}/{count} ["+"="*i+" "*(count-i)+"]")
            clear_output(wait=True)
            
        accuracy_mean/=count   
        ax[0].set_title(f"Epochs={epochs}\nBatch_size={batch_size}\nNumber of layers={number_of_layers[m]}\nNumber of neurons={number_of_neurons[n]}\nTotal neurons={total_neurons}\nAccuracy={accuracy_mean}")    
plt.show()
```

Мы видим, что для данной модели увеличение количества нейронов на слое примерно до количества входных нейронов позволяет увеличивать точность при обучение модели. При этом увеличение количества скрытых слоев увеличивает плотность точек на графиках, что говорит о том, что нейросеть обучается стабильнее и за меньшее число итераций находя больше зависимостей в данных. 

В результате симуляции различных архитектур нейросетей можно сделать вывод, что для данной задачи оптимальное решение- нейросеть имеющая 3 скрытых слоя по 20 нейронов. При данной конфигурации обучение происходит достаточно стабильно и быстро, при этом точность очень высокая в сравнение с другими вариантами архитектуры обученными в аналогичных условиях
